```python
# read in original data
data = pd.read_csv("ungroupedLabData_train.csv")
```

```python
labLabels = {"hemoglobin":"CBC", "hematocrit":"CBC","nucleated red cells":"CBC", "blood cells":"CBC" , "mch":"CBC", "mcv":"CBC", "platelet":"CBC", "rdw":"CBC", "rbc":"CBC", "wbc":"CBC",'poikilocytosis':"CBC", 'hypochromia':"CBC", "microcytes":"CBC", 'polychromasia':"CBC", 'anisocytosis':"CBC", 'macrocytes':"CBC", "monocyte":"CBC",
              "neutrophil":"CBC",   "eosinophil":"CBC", "basophil" :"CBC",
             "o2":"ABG", "base excess":"ABG", "oxygen":"ABG","alveolar-arterial":"ABG",
                 "calcium":"calcium",
                 "potassium":"Electrolytes", "chloride":"Electrolytes", "sodium":"Electrolytes", "bicarbonate":"Electrolytes","phosphate":"Electrolytes","magnesium":"Electrolytes",
                 "cholesterol":"cholesterol",
                 "glucose":"glucose",
                 "hepatitis":"hepatitis", "hiv":"HIV",
                 "bilirubin":"Bilirubin",
                 "ptt":"pt", "inr(pt)":"pt",
                 "aminotransferase":"aminotransferase",
                 "lactate":"lactate",
                 "lymph":"leukocytes", "eosinophils":"leukocytes", "monocytes":"leukocytes", "neutrophils":"leukocytes","basophils":"leukocytes", "granulocytes":"leukocytes", "metamyelocytes":"leukocytes", "myelocytes":"leukocytes",
                 "amylase":"pancreas", "lipase":"pancreas",
                 "creatine kinase":"creatine kinase", "ck-mb":"creatine kinase"
                }
labremove = ["top hold","bands","urine tube held","ventilat","waxy casts","edta hold" ,"peep","specimen","intubated","assist","tidal volume","tacrofk"]
```

```python
# remove infrequent events
data["label1"].value_counts()["glucose Blood"]
```

```python
events = list(set(data["label1"]))
```

```python
groupdic = {}
for e in events:
    print
    lowe = e.lower()
    if e in ["h Blood","i Blood","l Blood","bands Blood","sw  ","ns"] or lowe in labremove: continue
    
    if e == "ph Blood": 
        groupdic[e] = "ABG"
        continue
    if e == "pt Blood":
        groupdic[e] = "pt"
        continue
    
    if "Urine" in e: 
        groupdic[e] = "Urinalysis"
        continue
    if "Pleural" in e:
        groupdic[e] = "Pleural"
        continue
    if "Fluid"in e:
        groupdic[e] = "Fluid"
        continue
    if "Ascites" in e:
        groupdic[e] = "Ascites"
        continue
        
    for item in labLabels.keys():
        if item in lowe:
            groupdic[e] = labLabels[item]
            continue
    #print("here, ", e)
    if e not in groupdic.keys() and data["label1"].value_counts()[e] > 20:
        print(e)
        groupdic[e] = e
    
        
```

```python
set(groupdic.values())
```

```python
num = {}
i= 1
for item in set(groupdic.values()):
    num[item] = i
    i+=1
    
```

```python
group = []
for item in list(set(data["label1"])):
    if item in groupdic.keys():
        group.append([item, num[groupdic[item]]])
    else:
        group.append([item, 0])
```

```python
df = pd.DataFrame(group, columns=["event", "group"])
df    
```

```python
df[df["group"]==0]
```

```python
df.to_csv("expert.csv",index=False)
```
